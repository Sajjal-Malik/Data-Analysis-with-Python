# Plotting multiple Lines

```python
# Generate years array from 2000 to 2011
years = range(2000, 2012)

# Generate sample float values for apples and oranges yields
apples = [4.2, 4.5, 5.0, 5.5, 5.8, 6.0, 6.3, 6.7, 7.0, 7.2, 7.5, 7.8]
oranges = [3.0, 3.2, 3.8, 4.0, 4.5, 4.8, 5.0, 5.5, 5.8, 6.0, 6.3, 6.5]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')
plt.show()
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crop yeilds in Kanto Region')
plt.legend(['Apples', 'Oranges'])
plt.show()
```

# Markers and Grid

```python
# Plot apples and oranges yields
plt.plot(years, apples, marker='o', label='Apples')
plt.plot(years, oranges, marker='s', label='Oranges')

# Add title and labels
plt.title("Annual Fruit Yield (2000-2011)")
plt.xlabel("Year")
plt.ylabel("Yield (tons)")

# Show grid and legend
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

# Styling Lines and Markers

# Bar Chart

```python
# Data
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011]
apples = [10.5, 12.0, 13.5, 14.0, 15.5, 16.0, 17.0, 18.0, 19.0, 20.0, 21.0, 22.0]
oranges = [8.0, 9.5, 10.0, 11.0, 12.0, 13.5, 14.0, 15.0, 16.0, 17.0, 18.0, 19.0]

# Set the positions for the bars
x = np.arange(len(years))  # Positions for each year
width = 0.35  # Width of the bars

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Bar chart for apples and oranges
bars_apples = ax.bar(x - width/2, apples, width, label='Apples', color='lightgreen')
bars_oranges = ax.bar(x + width/2, oranges, width, label='Oranges', color='orange')

# Adding titles and labels
ax.set_title('Fruit Yield Over Years')
ax.set_xlabel('Year')
ax.set_ylabel('Yield (in tons)')
ax.set_xticks(x)
ax.set_xticklabels(years)
ax.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the flights dataset
flights_df = sns.load_dataset("flights")

# Pivot the data for the heatmap
flights_pivot = flights_df.pivot(index="month", columns="year", values="passengers")

# Create a grid for subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))  # 2 rows, 2 columns

# Line plot
sns.lineplot(data=flights_df, x="year", y="passengers", estimator="sum", ax=axes[0, 0])
axes[0, 0].set_title("Total Passengers Per Year")
axes[0, 0].set_xlabel("Year")
axes[0, 0].set_ylabel("Passengers")

# Heatmap
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="coolwarm", ax=axes[0, 1])
axes[0, 1].set_title("Monthly Passengers Heatmap")

# Bar plot
sns.barplot(data=flights_df, x="year", y="passengers", estimator="sum", ax=axes[1, 0])
axes[1, 0].set_title("Yearly Passenger Totals")
axes[1, 0].set_xlabel("Year")
axes[1, 0].set_ylabel("Total Passengers")

# Scatter plot
sns.scatterplot(data=flights_df, x="year", y="passengers", hue="month", ax=axes[1, 1])
axes[1, 1].set_title("Passenger Distribution Over Years")
axes[1, 1].legend(loc="upper left", bbox_to_anchor=(1, 1))  # Adjust legend position

# Adjust layout
plt.tight_layout()
plt.show()
```
